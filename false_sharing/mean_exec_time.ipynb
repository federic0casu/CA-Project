```python
import pandas as pd

images = [64, 256, 512, 1024, 2048]


for i in range(0,len(images)):
    for j in range(2,17):
        df____FS = pd.read_csv("csv/false_sharing/exec_times(" + str(images[i]) + "x" + str(images[i]) + "_4x4)(" + str(j) +").csv",header=None)
        df_NO_FS = pd.read_csv("csv/NO_false_sharing/exec_times(" + str(images[i]) + "x" + str(images[i]) + "_4x4)(" + str(j) +").csv",header=None)

        mean____FS = (df____FS[df____FS.columns[1]].mean()).round(10)
        mean_NO_FS = (df_NO_FS[df_NO_FS.columns[1]].mean()).round(10)
        speedup    = (mean____FS / mean_NO_FS).round(8)
        if (i<2 or (i==2 and j<10)):
            print("(" + str(images[i]) + "x" + str(images[i]) + ")(" + str(j) +")::mean    \t FS = " + str(mean____FS) + "\t|\t NO FS = " + str(mean_NO_FS) + "\t\t|\t RELATIVE SPEEDUP - " + str(speedup))
        else:
            print("(" + str(images[i]) + "x" + str(images[i]) + ")(" + str(j) +")::mean \t FS = " + str(mean____FS) + "\t|\t NO FS = " + str(mean_NO_FS) + "\t\t|\t RELATIVE SPEEDUP - " + str(speedup))    
```
